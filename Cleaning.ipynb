```python
# Step 2: Convert milliseconds to numeric again
results['milliseconds'] = pd.to_numeric(results['milliseconds'], errors='coerce')

# Step 3: Set a penalty value for NaNs in milliseconds
# This assumes milliseconds is a number representing total race time
max_millis = results['milliseconds'].max()
penalty_millis = max_millis + 10000  # Add 10 seconds in milliseconds

# Step 4: Fill NaNs with penalty time
results['milliseconds'] = results['milliseconds'].fillna(penalty_millis)

# Step 5: Convert milliseconds to time format in 'time' column
# Only update 'time' where it is currently NaN
def millis_to_time(ms):
    minutes = int(ms // 60000)
    seconds = (ms % 60000) / 1000
    return f"{minutes}:{seconds:06.3f}"

results['time'] = results.apply(
    lambda row: millis_to_time(row['milliseconds']) if pd.isna(row['time']) else row['time'],
    axis=1
)
```

```python
results['position'] = results['position'].fillna(99) 
results['rank'] = results['rank'].fillna(-1)
results['fastestLap'] = results['fastestLap'].fillna(0)
```

```python
# Helper: time string to seconds
def time_to_seconds(t):
    if pd.isna(t) or t == '00:00.000':
        return None
    try:
        m, s = t.split(':')
        return int(m) * 60 + float(s)
    except:
        return None

# Helper: seconds to time string
def seconds_to_time(s):
    minutes = int(s // 60)
    seconds = s % 60
    return f"{minutes}:{seconds:06.3f}"

# Sessions to process
sessions = ['fastestLapTime']

# Process each session
for session in sessions:
    sec_col = f'{session}_sec'

    # Step 1: Convert to seconds
    results[sec_col] = results[session].apply(time_to_seconds)

    # Step 2: Max value + 10s penalty
    max_val = results[sec_col].max()
    penalty_val = max_val + 10

    # Step 3: Fill NaNs safely (no inplace=True)
    results[sec_col] = results[sec_col].fillna(penalty_val)

    # Step 4: Convert back to time string
    results[session] = results[sec_col].apply(seconds_to_time)

    # Step 5: Drop the temp column
    results.drop(columns=[sec_col], inplace=True)
```

```python
# Helper: time string to seconds
def time_to_seconds(t):
    if pd.isna(t) or t == '00:00.000':
        return None
    try:
        m, s = t.split(':')
        return int(m) * 60 + float(s)
    except:
        return None

# Helper: seconds to time string
def seconds_to_time(s):
    minutes = int(s // 60)
    seconds = s % 60
    return f"{minutes}:{seconds:06.3f}"

# Sessions to process
sessions = ['q1', 'q2', 'q3']

# Process each session
for session in sessions:
    sec_col = f'{session}_sec'

    # Step 1: Convert to seconds
    qualifying[sec_col] = qualifying[session].apply(time_to_seconds)

    # Step 2: Max value + 10s penalty
    max_val = qualifying[sec_col].max()
    penalty_val = max_val + 10

    # Step 3: Fill NaNs safely (no inplace=True)
    qualifying[sec_col] = qualifying[sec_col].fillna(penalty_val)

    # Step 4: Convert back to time string
    qualifying[session] = qualifying[sec_col].apply(seconds_to_time)

    # Step 5: Drop the temp column
    qualifying.drop(columns=[sec_col], inplace=True)
```

```python
# ===== Rename Overlapping Columns =====
races.rename(columns={'name': 'race_name'}, inplace=True)
constructors.rename(columns={'name': 'constructor_name'}, inplace=True)

results.rename(columns={
    'time': 'result_time',
    'milliseconds': 'result_milliseconds',
    'position': 'final_position',
    'number': 'car_number'
}, inplace=True)

qualifying.rename(columns={
    'position': 'qualifying_position',
    'number': 'qualifying_car_number'
}, inplace=True)

lap_times.rename(columns={
    'time': 'lap_time',
    'milliseconds': 'lap_milliseconds',
    'position': 'lap_position'
}, inplace=True)

pit_stops.rename(columns={
    'time': 'pit_stop_time',
    'milliseconds': 'pit_stop_milliseconds',
    'duration': 'pit_stop_duration'
}, inplace=True)

# ===== Merge Base Tables =====
finalF1 = results.merge(races, on='raceId', how='left') \
                   .merge(drivers, on='driverId', how='left') \
                   .merge(constructors, on='constructorId', how='left') \
                   .merge(status, on='statusId', how='left')

# ===== Merge Qualifying (simplified to avoid car_number mismatches) =====
finalF1 = finalF1.merge(
    qualifying,
    on=['raceId', 'driverId'],
    how='left',
    suffixes=('', '_qualifying')
)

# ===== Aggregate Lap Times =====
lap_summary = lap_times.groupby(['raceId', 'driverId']).agg({
    'lap_position': 'mean',
    'lap_time': 'count'
}).rename(columns={
    'lap_position': 'avg_lap_position',
    'lap_time': 'total_laps'
}).reset_index()

# ===== Aggregate Pit Stops =====
pit_summary = pit_stops.groupby(['raceId', 'driverId']).agg({
    'stop': 'count',
    'pit_stop_duration': 'sum'
}).rename(columns={
    'stop': 'num_pit_stops',
    'pit_stop_duration': 'total_pit_stop_duration'
}).reset_index()

# ===== Final Merge =====
finalF1 = finalF1.merge(lap_summary, on=['raceId', 'driverId'], how='left')
finalF1 = finalF1.merge(pit_summary, on=['raceId', 'driverId'], how='left')

# ===== Optional: Save Final Dataset =====

```
